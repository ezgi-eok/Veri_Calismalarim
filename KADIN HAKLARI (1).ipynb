## KADIN HAKLARI

1. Female to Male Ratio of Time Devoted to Unpaid Care Work (OECD):
-Problem Tanımı: Unpaid care work, özellikle kadınların üzerine yüklenen, ekonomik değeri olmasına rağmen genellikle görünmeyen bir iştir. OECD verileri üzerinden kadınların erkeklere oranla daha fazla zaman harcadığı bu alandaki eşitsizlik, toplumsal cinsiyet eşitliği çabalarına zarar vermektedir.
-Veri Seti Bilgisi: OECD ülkelerinin yıllık verileri, kadınların ve erkeklerin toplam çalışma süreleri içinde ne kadarını ücretsiz bakım işlerine ayırdığını gösteren oranlar.
-Veri Analizi: Ülkeler arasında karşılaştırmalar, zaman içindeki değişim trendleri, gelir düzeyi ile oranın ilişkisi gibi analizler.
-Bulgular ve Görüşler: Kadınların hala erkeklerden çok daha fazla zaman harcadığı görülmektedir. Toplumsal cinsiyet eşitliğini artırmak için, erkeklerin de bu alandaki sorumluluğu paylaşması gerektiği vurgulanabilir.
-Sonuç: Daha fazla politika müdahalesi ve eğitim gereklidir, erkeklerin ev içindeki iş gücüne katılımı teşvik edilmelidir.

2. Share of Women in Top 0.1%, 0.25%, 0.5%, 1%, 5%, and 10%:
-Problem Tanımı: Kadınların üst düzey gelir gruplarında ne kadar yer aldığı, ekonomik eşitsizliklerin bir göstergesidir. Kadınların bu grupta yer alma oranlarının düşük olması, toplumsal cinsiyet eşitsizliğinin ekonomi üzerindeki etkilerini ortaya koymaktadır.
-Veri Seti Bilgisi: Ülkelerin ekonomik verileri üzerinden, farklı gelir dilimlerinde kadınların payı.
-Veri Analizi: Kadınların gelir dağılımındaki yerini göstermek için istatistiksel analizler ve farklı ülkeler arasındaki karşılaştırmalar yapılabilir.
-Bulgular ve Görüşler: Kadınların yüksek gelirli gruplarda erkeklere göre belirgin şekilde daha az yer aldığı ve buna neden olan yapısal engellerin bulunduğu sonucuna varılabilir.
-Sonuç: Kadınların üst gelir dilimlerinde daha fazla yer alabilmesi için daha fazla fırsat eşitliği sağlanmalıdır.

3. Ratio of Female to Male Labor Force Participation Rate:
-Problem Tanımı: Kadınların iş gücüne katılım oranı, ekonomik eşitsizlikleri ve toplumsal cinsiyet eşitliğini gösteren önemli bir göstergedir. Erkeklerin iş gücüne katılımı ile kadınların katılımı arasındaki fark, eşitsizlikleri ortaya koyar.
-Veri Seti Bilgisi: İş gücüne katılım oranları, cinsiyete göre ayrılmış veriler.
-Veri Analizi: Kadın ve erkeklerin iş gücüne katılım oranları arasındaki farklar, sektörel bazda analizler.
-Bulgular ve Görüşler: Çalışma gücüne katılım oranları arasındaki farklar, kadınların ekonomik bağımsızlık ve kariyer fırsatlarında yaşadıkları eşitsizlikleri ortaya koymaktadır.
-Sonuç: Kadınların iş gücünü katılımını artırmak için daha fazla destek ve esnek çalışma modelleri gereklidir.

4. Maternal Mortality Ratio:
-Problem Tanımı: Annelik sırasında ölüm oranları, bir ülkenin sağlık sisteminin etkinliğini ve kadın hakları alanındaki eksiklikleri gösteren kritik bir göstergedir. Kadınların sağlık hizmetlerine erişimindeki eşitsizlikler, annelik ölüm oranlarını artırabilir.
-Veri Seti Bilgisi: Ülkeler arasındaki annelik ölüm oranlarına dair veriler.
-Veri Analizi: Annelik ölüm oranlarını göstermek için ülke bazlı karşılaştırmalar ve risk faktörlerinin analiz edilmesi.
-Bulgular ve Görüşler: Yoksul ve kırsal alanlarda kadınlar, kaliteli sağlık hizmetlerine ulaşmada zorluk çekiyor. Bu durum, annelik ölüm oranlarını yükseltmektedir.
-Sonuç: Sağlık hizmetlerine eşit erişimi sağlamak için politika reformları yapılmalıdır.

5. Gender Wage Gap:
-Problem Tanımı: Cinsiyetler arasındaki maaş farkı, kadınların iş gücündeki durumunu gösteren önemli bir göstergedir. Kadınlar aynı işleri yapmalarına rağmen erkeklerden daha düşük maaş alabiliyorlar.
-Veri Seti Bilgisi: Ülkeler arasındaki cinsiyetler arası maaş farklarını gösteren veriler.
-Veri Analizi: Farklı sektörlerdeki maaş farkları, eğitim düzeyine göre farklar, devlet politikaları ile ilişkilendirme.
-Bulgular ve Görüşler: Kadınların, erkeklerden düşük maaşlar aldıkları sektörler ve pozisyonlar belirgin olarak gözlemlenebilir.
-Sonuç: Kadınların eşit maaş alabilmesi için işe alım süreçlerinde şeffaflık ve cinsiyet ayrımcılığını ortadan kaldıran politikalar gereklidir.
